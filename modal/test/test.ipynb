```python
import requests
import jwt
import time
import os
from IPython.display import Video

# Replace with your actual endpoint and API key
ENDPOINT = "https://your-endpoint-url.modal.run/generate"
API_KEY = "aibuttons123"

# Create JWT token
def create_token(api_key):
    payload = {
        "exp": int(time.time()) + 3600,  # Token expires in 1 hour
        "iat": int(time.time())
    }
    return jwt.encode(payload, api_key, algorithm="HS256")

# Function to call the API and save the video
def generate_video(image_url, prompt, output_file="output.mp4"):
    token = create_token(API_KEY)
    
    headers = {
        "Authorization": f"Bearer {token}"
    }
    
    data = {
        "image_url": image_url,
        "prompt": prompt,
        # Default parameters below - adjust as needed
        "height": 720,
        "width": 1280,
        "num_frames": 45,
        "fps": 24
    }
    
    print(f"Sending request to generate video from image: {image_url}")
    response = requests.post(ENDPOINT, headers=headers, json=data, stream=True)
    
    if response.status_code == 200:
        with open(output_file, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"Video saved to {output_file}")
        return output_file
    else:
        print(f"Error: {response.status_code}")
        print(response.text)
        return None

# Example usage
image_url = "https://github.com/yavuzceliker/sample-images/blob/main/images/image-100.jpg?raw=true"  # Replace with your image URL
prompt = "A camera panning through a beautiful forest"
output_file = "hunyuan_video.mp4"

video_path = generate_video(image_url, prompt, output_file)

if video_path:
    # Display the video in the notebook
    display(Video(video_path, embed=True))
```
